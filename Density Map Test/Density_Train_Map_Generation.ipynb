```python
import matplotlib.pyplot as plt
import numpy
import json
import PIL
from PIL import ImageDraw
```

```python
data = None

# Load the json
with open("0009.json") as f:
  data = json.load(f)

# Get the array of points
image_data = data["points"]
```

```python
image_load = PIL.Image.open("0009.jpg")

width, height = image_load.size
```

```python
img = PIL.Image.new("RGB", (width, height))
draw = ImageDraw.Draw(img)

for points in image_data:
    draw.ellipse((points[0] - 50, points[1] - 50, points[0] + 50, points[1] + 50), fill=(0, 255, 0), outline=(0, 0, 0))

plt.imshow(img)
```

```python
image = PIL.Image.open("0009.jpg")
draw = ImageDraw.Draw(image)

for points in image_data:
    draw.ellipse((points[0] - 50, points[1] - 50, points[0] + 50, points[1] + 50), fill=(0, 255, 0), outline=(0, 0, 0))

plt.imshow(image)
```

```python
for i in range(len(image_data)):
    image_data[i][0] = image_data[i][0] / width * 255
    image_data[i][1] = image_data[i][1] / height * 255
```

```python
img = PIL.Image.new("RGB", (255, 255))
draw = ImageDraw.Draw(img)

for points in image_data:
    draw.ellipse((points[0] - 2, points[1] - 2, points[0] + 2, points[1] + 2), fill=(0, 255, 0), outline=(0, 0, 0))

plt.imshow(img)
```

```python
image_resize = PIL.Image.open("0009.jpg")
image_resize = image_resize.resize((255, 255))

draw = ImageDraw.Draw(image_resize)

for points in image_data:
    draw.ellipse((points[0] - 2, points[1] - 2, points[0] + 2, points[1] + 2), fill=(0, 255, 0), outline=(0, 0, 0))

plt.imshow(image_resize)
```
